##VP


```python
#Valor presente
def fun_valor_presente(valor_futuro,taxa,tempo):
  valor_presente = valor_futuro/(1+taxa)**tempo
  return (valor_presente)
```

##VPL

```python
#vai trazer uma serie de pagamentos a valor presente

valor_pago_icial = -155000
valor_pagoa_e_ganho = 70000-15000
fc = np.array([valor_pago_icial, valor_pagoa_e_ganho, valor_pagoa_e_ganho,  valor_pagoa_e_ganho, valor_pagoa_e_ganho, valor_pagoa_e_ganho, valor_pagoa_e_ganho])

investimento = np.npv(rate=0.17,values=fc)

print("VPL", investimento)
```

##TIR


```python
#demonstra a taxa que rende um projeto de investimentos considerando a periodicidade dos fluxos de caixas do projeto

fluxo =np.zeros(7)
fluxo[0] = -155000
for i in range(6):
  fluxo[i+1] = 70000-15000 #ganha R$7000 e gasta R$1500

tir = np.irr(fluxo)#o calculo ja é nativo da numpy

print('Tir : '+ str(round(100*tir,2)) +'%')
```
