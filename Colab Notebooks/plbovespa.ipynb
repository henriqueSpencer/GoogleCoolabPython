```python
#import pandas as pd

#df = pd.read_csv("/Users/daniel/Documents/Python/python_files/plbovespa.csv", sep = ';', header = 0)
df.columns=['Date', 'Close']

#calculo
df['SMA20'] = df['Close'].rolling(20).mean()
df['Upper'] = df['SMA20'] + 2*(df['Close'].rolling(20).std())
df['Lower'] = df['SMA20'] - 2*(df['Close'].rolling(20).std())
```

```python
#plotting
import matplotlib.pyplot as plt

df[['PL', 'SMA20', 'Upper', 'Lower']].tail(1ooo).plot(figsize=(16,6))
plt.xlabel('Data')
plt.ylabel('Preco/Lucro')
plt.title('PL BOVESPA')
plt.legend(loc = 'lower left')
```
