#Fazendo onda qudrada


```python
x_quadrada = linspace(0,10,110)
x_quadrada
```

```python
y_quadrada=[0]*110
```

```python
for i in range(110):
  if (0<x_quadrada[i]<1):
    y_quadrada[i]=-0.5
  if (1<x_quadrada[i]<2):
    y_quadrada[i]=0.5  
  if (2<x_quadrada[i]<3):
    y_quadrada[i]=-0.5
  if (3<x_quadrada[i]<4):
    y_quadrada[i]=0.5  
  if (4<x_quadrada[i]<5):
    y_quadrada[i]=-0.5
  if (5<x_quadrada[i]<6):
    y_quadrada[i]=0.5
  if (6<x_quadrada[i]<7):
    y_quadrada[i]=-0.5  
  if (7<x_quadrada[i]<8):
    y_quadrada[i]=0.5
  if (8<x_quadrada[i]<9):
    y_quadrada[i]=-0.5  
  if (9<x_quadrada[i]<10):
    y_quadrada[i]=0.5 
```

```python
y_quadrada
```

```python
plot(x_quadrada,y_quadrada)

plt.grid(True)
plt.title(r'Onda Quadrada com 1Vpp ', fontsize=20)
plt.xlabel('tempo (s)')
plt.ylabel('volts (V)')
plt.show()
```

```python
y2_quadrada=[0]*110
```

```python

for i in range(110):
    y2_quadrada[i]=2*y_quadrada[i]

plot(x_quadrada,y2_quadrada)

plt.grid(True)
plt.title(r'Soma de duas ondas quadradas de 1Vpp cada ', fontsize=20)
plt.xlabel('tempo (s)')
plt.ylabel('volts (V)')
plt.show()
```

```python
plot(x_quadrada+x_senoide , y_senoide+y_quadrada)

plt.grid(True)
plt.title(r'Senoide + Onda Quadrada ', fontsize=20)
plt.xlabel('tempo (s)')
plt.ylabel('volts (V)')
plt.show()
```

##Onda triangular


```python
x_triangular = linspace(0,10,11)
x_triangular
```

```python
#y_quadrada=[0]*110
y_triangular=[0]*11
for i in range(11):
  y_triangular[i]=(-1)**i

plot (x_triangular,y_triangular)
plt.grid(True)
plt.title(r'Onda triangular', fontsize=20)
plt.xlabel('tempo (s)')
plt.ylabel('volts (V)')
plt.show()
```
